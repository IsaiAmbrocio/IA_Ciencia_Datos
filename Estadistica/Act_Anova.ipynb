# Anova unifactorial

El anova se utiliza para comparar más de dos medias poblacionales

 $$ H_0: \mu_1 = \mu_2 = \cdots = \mu_i$$

 contra $H_a:$  al menos alguna $\mu_i$ es diferente.
 El estadístico de prueba tiene distribución $F$

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.stats.multicomp as multi
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv("/content/Iris.csv")
```

```python
df.head()
```

# Anova usando el factor sepalo.

```python
longitud_especie = ols("SepalLengthCm~C(Species)", data = df).fit()
sm.stats.anova_lm(longitud_especie, type = 2)
```

Dado que el P-Valor `PR(>F)`= 1.669669e-31 < 0.05
Se rechaza $H_0: \mu_1 = \mu_2 = \mu_3$ es decir al menos un $\mu_i$ es diferente.

```python
modelo = ols(formula = "SepalLengthCm~C(Species)", data = df)
modelo.fit().summary()
```

# Utilizar test Tukey (Método T) para saber cual especie es diferente.

```python
multi_comp = multi.MultiComparison(df["SepalLengthCm"], df["Species"])
multi_comp.tukeyhsd().summary()
```
